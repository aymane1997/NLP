```python
import json
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lr_loss = {}
negative_samples_loss=  {}
windowsize_loss = {}
```

```python
lr_list = [0.005,0.01,0.05,0.1, 0.2]
neg_sam_list = [5,10,15,20,25]
win_size_list = [5,10,15,20,25]
```

```python
for lr in lr_list:
    with open(f'mymodel.model_lr = {lr}') as fp:
        lr_loss[lr] = np.array(json.load(fp)['loss'])
fig, ax = plt.subplots(figsize = (10,10))
pd.DataFrame.from_dict(lr_loss).add_prefix('lr = ').plot(ax = ax, color = 'brgkm', alpha = 0.7)
ax.set_ylabel('loss')
ax.set_title('loss values by learning rate')
plt.show()
```

```python
for neg in neg_sam_list:
    with open(f'mymodel.model_negative_rate = {neg}') as fp:
        negative_samples_loss[neg] = np.array(json.load(fp)['loss'])
fig, ax = plt.subplots(figsize = (10,10))
pd.DataFrame.from_dict(negative_samples_loss).add_prefix('negative Rate = ').plot(ax = ax, color = 'brgkm', alpha = 0.7)
ax.set_ylabel('loss')
ax.set_title('loss values by negative rates')
plt.show()
```

```python
for win in win_size_list:
    with open(f'mymodel.model_window = {win}') as fp:
        windowsize_loss[win] = np.array(json.load(fp)['loss'])
fig, ax = plt.subplots(figsize = (10,10))
pd.DataFrame.from_dict(negative_samples_loss).add_prefix('window_size = ').plot(ax = ax, color = 'brgkm', alpha = 0.7)
ax.set_ylabel('loss')
ax.set_title('loss values by window sizes')
plt.show()
```
